# Import the image
Define a function to which we pass the name of our image (assuming its in the ./image directory) as well as the number of colors we want to have in our resulting image. The function returns the image with the pixel values adjusted to be one of the selected colors using a KMeans approach.

```python
def reduce_image(filename, k):
    # Load the image from file and reformat
    image = np.array(cv2.cvtColor(cv2.imread("images/" + filename), cv2.COLOR_BGR2RGB))

    # Provide a way to simply fetch image from filename
    if k == -1:
        return image, None, None
    
    dims = image.shape
    image = image.reshape(dims[0]*dims[1], 3)
    
    # Find cluster centers
    kmeans = KMeans(random_state=101, n_clusters=k, init='random', n_init="auto").fit(image)
    
    # For each pixel, find nearest cluster and set pixel to cluster center value
    assignments = kmeans.predict(image)
    for i in range(dims[0]*dims[1]):
        image[i] = kmeans.cluster_centers_[assignments[i]]
        
    # Return the updated image 
    return image.reshape(dims), assignments, kmeans.cluster_centers_
```

# Create simple coordinate object
Create a simple object that represents a coordinate in the 2D image array. Just simplifies the contour-finding operations that will be performed later

```python
class Coord:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Region:
    def __init__(self, color):
        self.color = color
        self.coordinates = []
    def addEdge(self, coord):
        self.coordinates.append(coord)
```

```python
def pixel_equal(pix_1, pix_2):
    if pix_1[0] != pix_2[0]:
        return False
    if pix_1[1] != pix_2[1]:
        return False
    if pix_1[2] != pix_2[2]:
        return False
    return True

def coord_to_str(coord):
    return f"{coord.x},{coord.y}"
    
def str_to_coord(string):
    first, second = string.split(",")
    return Coord(int(first), int(second))
```

# Contour finding function
Define a function that accepts a 3D array representing an image (two dimensional grid of pixels, and each pixel has 3 colors in RGB format) and finds the contours around pixels of the same color. Returns an array of pixel coords that are the contours of the image.

```python
def find_contours(image):
    # Find first blob
    #horizon = deque()
    contour_x = []
    contour_y = []
    for x in range(0, len(image)):
        for y in range(0, len(image[x])):
            # Check left neighbor
            if x > 0:
                if not pixel_equal(image[x][y], image[x-1][y]):
                    contour_x.append(x)
                    contour_y.append(y)
                    continue
            # Check right neighbor
            if x < len(image) - 1:
                if not pixel_equal(image[x][y], image[x+1][y]):
                    contour_x.append(x)
                    contour_y.append(y)
                    continue
            # Check bottom neighbor
            if y > 0:
                if not pixel_equal(image[x][y], image[x][y-1]):
                    contour_x.append(x)
                    contour_y.append(y)
                    continue
            # Check top neighbor
            if y < len(image) - 1:
                if not pixel_equal(image[x][y], image[x][y+1]):
                    contour_x.append(x)
                    contour_y.append(y)
                    continue
    return contour_x, contour_y
```

```python
def discover_neighbors(image, curCoord, visited):
    cx = curCoord.x
    cy = curCoord.y
    valid_neighbors = []
    # Check x+1 coord first
    if cx < len(image) - 1:
        if pixel_equal(image[cx][cy], image[cx+1][cy]):
            if visited[cx+1][cy] == 0:
                valid_neighbors.append(Coord(cx+1, cy))
                visited[cx+1][cy] = 1
    # Check x-1
    if cx > 0:
        if pixel_equal(image[cx][cy], image[cx-1][cy]):
            if visited[cx-1][cy] == 0:
                valid_neighbors.append(Coord(cx-1, cy))
                visited[cx-1][cy] = 1
    # Check y+1
    if cy < len(image[0]) - 1:
        if pixel_equal(image[cx][cy], image[cx][cy+1]):
            if visited[cx][cy+1] == 0:
                valid_neighbors.append(Coord(cx, cy+1))
                visited[cx][cy+1] = 1
    # Check y-1
    if cy > 0:
        if pixel_equal(image[cx][cy], image[cx][cy-1]):
            if visited[cx][cy-1] == 0:
                valid_neighbors.append(Coord(cx, cy-1))
                visited[cx][cy-1] = 1

    #print(f"valid neighbors has {len(valid_neighbors)} elements")
    return valid_neighbors

def is_edge(image, curCoord):
    cx = curCoord.x
    cy = curCoord.y
    causal_neighbors = set()
    edge_flag = False
    # Pixel exists on left or right image boundary
    if cx == 0 or cx == (len(image) - 1):
        edge_flag = True
    # Pixel exists on top or bottom image boundary
    if cy == 0 or cy == (len(image[0]) - 1):
        edge_flag = True
    # Pixel has different color than x+1 neighbor
    if (cx + 1)  <= (len(image) - 1):
        if not pixel_equal(image[cx][cy], image[cx+1][cy]):
            edge_flag = True
            causal_neighbors.add(coord_to_str(Coord(cx+1, cy)))
    # Pixel has different color than x-1 neighbor
    if (cx - 1)  >= 0:
        if not pixel_equal(image[cx][cy], image[cx-1][cy]):
            edge_flag = True
            causal_neighbors.add(coord_to_str(Coord(cx-1, cy)))
    # Pixel has different color than y+1 neighbor
    if (cy + 1)  <= (len(image[0]) - 1):
        if not pixel_equal(image[cx][cy], image[cx][cy+1]):
            edge_flag = True
            causal_neighbors.add(coord_to_str(Coord(cx, cy+1)))
    # Pixel has different color than y-1 neighbor
    if (cy - 1)  >= 0:
        if not pixel_equal(image[cx][cy], image[cx][cy-1]):
            edge_flag = True
            causal_neighbors.add(coord_to_str(Coord(cx, cy-1)))
            
    return edge_flag, causal_neighbors
    
def find_contours_v2(image):
    region_size_threshold = 2
    regions = {}
    visited = np.array([[0 for i in range(len(image[0]))] for j in range(len(image))])
    #print(visited.shape)
    #print(image.shape)
    ex_frontier = set()
    in_frontier = []
    ex_frontier.add(coord_to_str(Coord(0,0)))
    visited[0][0] = 1
    
    while len(ex_frontier) > 0 and len(regions) < 10000:
        # Setup for discovering new region
        #print("top of loop")
        pixStr = ex_frontier.pop()
        firstPixel = str_to_coord(pixStr)
        curRegion = Region(image[firstPixel.x][firstPixel.y])
        # Setup frontier for new region
        in_frontier.append(firstPixel)

        # Perform the discovery steps
        while len(in_frontier) > 0:
            # Grab a pixel
            curPixel = in_frontier.pop()
            # Check if is an edge and take appropriate action
            validEdge, causalNeighbors = is_edge(image, curPixel)
            if validEdge:
                curRegion.addEdge(curPixel)
                #print("found valid edge")
                for neigh in causalNeighbors:
                    coord = str_to_coord(neigh)
                    if visited[coord.x][coord.y] == 0:
                        ex_frontier.add(neigh)
                        visited[coord.x][coord.y] = 1
            # Add valid neighbors to frontier
            nextNeighbors = discover_neighbors(image, curPixel, visited)
            for neigh in nextNeighbors:
                neighStr = coord_to_str(neigh)
                ex_frontier.discard(neighStr)
            in_frontier.extend(nextNeighbors)
            #print("adding neighbors")

        # Add new region to growing collection
        if len(curRegion.coordinates) < region_size_threshold:
            for coord in curRegion.coordinates:
                ex_frontier.discard(coord_to_str(coord))
        else:
            regions[len(regions)] = curRegion
        #print(len(ex_frontier))
        #print("added new region")
    return regions, visited
                
```

Perform some processing to regions so that the borders can be drawn by continuous lines instead of by pixels

```python
def mass_coord_to_str(coordinates):
    str_coords = set()
    for coord in coordinates:
        str_coords.add(coord_to_str(coord))
        print(coord.x, coord.y)
    return str_coords
```

```python
def order_border(str_coords):
    # Setup
    ordered_x, ordered_y = ([], [])
    prepend_x, prepend_y = ([], [])
    single_fail = False
    
    # Get a starting point in the region's boundary
    first_point_str = str_coords.pop()
    first_point_coord = str_to_coord(first_point_str)
    
    cur_point_coord = first_point_coord
    ordered_x.append(cur_point_coord.x)
    ordered_y.append(cur_point_coord.y)
    
    
    # Continuously select adjacent points to add to new list
    while len(str_coords) > 0:
        possible_coords = [Coord(cur_point_coord.x+1, cur_point_coord.y),
                           Coord(cur_point_coord.x-1, cur_point_coord.y),
                           Coord(cur_point_coord.x, cur_point_coord.y+1),
                           Coord(cur_point_coord.x, cur_point_coord.y-1)]
        found = False
        for next_point_coord in possible_coords:
            next_point_str = coord_to_str(next_point_coord)
            if next_point_str in str_coords:
                str_coords.remove(next_point_str)
                if single_fail == False:
                    ordered_x.append(next_point_coord.x)
                    ordered_y.append(next_point_coord.y)
                else:
                    prepend_x.append(next_point_coord.x)
                    prepend_y.append(next_point_coord.y)
                cur_point_coord = next_point_coord
                found=True
                break  
        # If we cannot find any more points along edge, try going other direction for starting point
        if not found:
            if single_fail == False:
                cur_point_coord = first_point_coord
            else:
                break;

    return prepend_x + ordered_x, prepend_y + ordered_y
# Must keep in mind notion of direction, since line should tend to continue in same direction if possible
```
